```python
import numpy as np
from matplotlib import pyplot as plt 
#matplotlib inline
```

## subplot example

```python
x1 = np.random.normal(1,0.5, 20)
y1 = np.arange(0,len(x1),1)

x2 = np.random.normal(10,0.5 , 20)
y2 = np.arange(0,len(x1),1)
```

+ 1x1 subplot

```python
fig , ax = plt.subplots()
ax.plot(x1,y1)
ax.set_title("single plot")
```

+ N x 1 , 1 x N subplot

```python
fig , ax = plt.subplots( 1 , 2 , sharey = True , sharex = True)
# https://bioboot.github.io/web-2016/class-material/day2-visualization.pdf

ax[0].scatter(x1,y1)
ax[1].scatter(x2,y2,color='r')
```
